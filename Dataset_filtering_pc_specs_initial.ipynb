## **Function to create a dateframe indicating count of unique values**

```python
def unique_chart(Dataset):
  uni_dict = {}
  uni_value ={}
  keys = Dataset.columns
  for i in keys:
    uni_dict[i] = Dataset[i].nunique()
    if Dataset[i].nunique() <= 10:
      uni_value[i] = Dataset[i].unique()
    else:
      uni_value[i] = None
  return (uni_dict, uni_value)
```

```python
uni_dict, uni_value = unique_chart(pc_details)
Unique = pd.DataFrame([uni_dict, uni_value]) 
print(Unique.to_markdown())
```

```python
Unique.columns
```

Observation: From the above operations, we can see that some of the columns have high number of null anf unique values that will not contribute towards further analysis. So, we can drop them.
