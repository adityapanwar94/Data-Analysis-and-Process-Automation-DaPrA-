```python
columns = ['Memory1', 'Memory2', 'Memory3', 'Memory4']

def Memory_replace(columns):  # Removing text data
  for cols in columns:
    pc_details[cols] = pc_details[cols].str.split(' ').str[0]
  return pc_details

pc_details = Memory_replace(columns)
```

```python
a=  pc_details['Memory1'].unique()
print(a)
```

```python
columns = ['Memory1', 'Memory2', 'Memory3', 'Memory4']

def Memory_null_replace(columns):   # Converting nil and null values to 0
  for cols in columns:
    pc_details[cols] = pc_details[cols].replace('Nil' , '')
    pc_details[cols] = pc_details[cols].replace(r'^\s*$', np.nan, regex=True)
    pc_details[cols] = pc_details[cols].fillna(0)
  return pc_details

pc_details = Memory_null_replace(columns)
```

## **Cleaning Graphic card details**

```python
columns = ['Graphic Card1', 'Graphic Card2', 'Graphic Card3', 'Graphic Card4']

def Graphic_card_replace(columns):  # Removing text data
  for cols in columns:
    pc_details[cols] = pc_details[cols].str.split(' ').str[0]
  return pc_details

pc_details = Graphic_card_replace(columns)
```

```python
a = pc_details['Graphic Card1'].unique()
print(a) 
```

```python
columns = ['Graphic Card1', 'Graphic Card2', 'Graphic Card3', 'Graphic Card4']

def Graphic_null_replace(columns):   # Converting nil and null values to 0 and replacing text data
  for cols in columns:
    pc_details[cols] = pc_details[cols].replace('Nil' , '')
    pc_details[cols] = pc_details[cols].apply(pd.to_numeric,errors='coerce') # Replace text data with NaN
    pc_details[cols] = pc_details[cols].fillna(0)
  return pc_details

pc_details = Graphic_null_replace(columns)
```

## **Cleaning opencl score data**

```python
a = pc_details['OpenCL score1'].unique()
print(a) 
```

```python
columns = ['OpenCL score1', 'OpenCL score2', 'OpenCL score3', 'OpenCL score4']

def Opencl_clean(columns):   # Converting nil and null values to 0 and replacing text data
  for cols in columns:
    pc_details[cols] = pc_details[cols].replace('Nil' , '')
    pc_details[cols] = pc_details[cols].apply(pd.to_numeric,errors='coerce') # Replace text data with NaN
    pc_details[cols] = pc_details[cols].fillna(0)
  return pc_details

pc_details = Opencl_clean(columns)
```
