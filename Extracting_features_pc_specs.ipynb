## **Import Libraries**

```python
import pandas as pd
import numpy as np
```

## **Read csv file**

```python
pc_details = pd.read_csv(r'/content/drive/MyDrive/dataset/Cleaned_pc_specs2.csv')
pc_details.info()
```

```python
pc_details.head()
```

## **Dropping unwanted columns**

```python
pc_details.columns
```

```python
pc_details = pc_details.drop(['Unnamed: 0.1', 'OS1', 'OS2', 'OS3', 'OS4', 'Model1', 'Model2', 'Model3', 'Model4', 
                              'CPU1', 'CPU2', 'CPU3', 'CPU4', 'OpenCL score1', 'OpenCL score2', 'OpenCL score3','OpenCL score4', 
                              'Graphic Card1', 'Graphic Card2', 'Graphic Card3', 'Graphic Card4'], axis=1)
```

```python
pc_details.head()
```

## **Getting mean value of single core score**

```python
pc_details = pc_details.assign(single_core_avg=pc_details.loc[:, ['Single core score1',
       'Single core score2', 'Single core score3', 'Single core score4']].mean(axis=1))
```

```python
pc_details.head()
```

## **Getting mean value of multi core score**

```python
pc_details = pc_details.assign(multi_core_avg=pc_details.loc[:, ['Multi core score1', 'Multi core score2', 'Multi core score3',
       'Multi core score4']].mean(axis=1))
```

```python
pc_details.head()
```

## **Getting average of memory column**

```python
pc_details = pc_details.assign(memory_avg=pc_details.loc[:, ['Memory1','Memory2', 'Memory3', 'Memory4']].mean(axis=1))
```

```python
pc_details.head()
```

```python
pc_details.info()
```

## **Creating a cloumn for geekbench score**

```python
pc_details = pc_details.assign(Geekbench_score = pc_details.loc[:, ['single_core_avg', 'multi_core_avg']].mean(axis=1))
```

```python
pc_details.head()
```

```python
pc_details.columns
```

```python
pc_details1 = pc_details.drop(['Single core score1', 'Single core score2',
       'Single core score3', 'Single core score4', 'Multi core score1',
       'Multi core score2', 'Multi core score3', 'Multi core score4',
       'Memory1', 'Memory2', 'Memory3', 'Memory4'], axis=1)
```

```python
pc_details1.info()
```

```python
pc_details1.head()
```

```python
pc_details1.to_csv('pc-specs_final.csv', index = False)
```
